```python
import cv2
import numpy as np
import h5py
from matplotlib import pyplot as plt
```

```python
def visualize(**images):
    """PLot images in one row."""
    n = len(images)
    plt.figure(figsize=(16, 5))
    for i, (name, image) in enumerate(images.items()):
        plt.subplot(1, n, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.title(' '.join(name.split('_')).title())
        plt.imshow(image, cmap='gray')
        
    plt.tight_layout()
    plt.show()
```

```python
files = h5py.File('input/train_mscale.h5')
```

```python
images = files['data'][:]
labels = files['label'][:]
```

```python
for index in np.random.randint(0, images.shape[0] - 1, (5,)):
    img = (images[index] * 255).astype(np.uint8)
    img = np.squeeze(img, axis=0)
    
    img_label = (labels[index] * 255).astype(np.uint8)
    img_label = np.squeeze(img_label, axis=0)
    
    visualize(
        blur_img=img, clear_img=img_label
    )
```
